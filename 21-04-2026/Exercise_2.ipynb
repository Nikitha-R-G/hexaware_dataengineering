```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def sale_category(x):
    if x>50000:
        return "Premium"
    elif x>=20000:
        return "Medium"
    else:
        return "Small"

udf_func = udf(sale_category,StringType())

sales.withColumn(
"sale_category",
udf_func("amount")
).show()
```

```python
import ipywidgets as widgets
```

```python
dept = widgets.Dropdown(options=["IT","HR","Finance"])
display(dept)
```

```python
salary_slider = widgets.IntSlider(min=50000,max=80000)
display(salary_slider)
```

```python
employees.filter(
(col("department")==dept.value) &
(col("salary")>=salary_slider.value)
).show()
```

```python
top_employee = employees.orderBy(col("salary").desc()).first()

avg_salary = employees.select(avg("salary")).first()[0]

dept_count = employees.groupBy("department").count()

total_sales = sales.select(sum("amount")).first()[0]

best_sales_emp = sales.groupBy("emp_id") \
.sum("amount") \
.orderBy(col("sum(amount)").desc()) \
.first()
```

```python
with open("final_report.txt","w") as f:
    f.write(f"Top Employee: {top_employee}\n")
    f.write(f"Average Salary: {avg_salary}\n")
    f.write(f"Total Sales: {total_sales}\n")
    f.write(f"Best Sales Employee: {best_sales_emp}\n")
```
