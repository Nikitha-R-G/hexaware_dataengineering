```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("app Basics").getOrCreate()
print("SparkSssion created")
```

```python
data=[
    ("Niki",600000,"CSE"),
    ("Praki",800000,"ECE"),
    ("Tharun",1000000,"CSE"),
    ("Sarvesh",500000,"IT"),
    ("Sanjeev",800000,"Mech")
]
columns=["Name","Salary","Department"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.select("Name","Salary").show()
```

```python
df.filter(df.Salary>50000).show()
```

```python
df.groupBy("Department").count().show()
```

```python
df.orderBy("Salary").show()
```
