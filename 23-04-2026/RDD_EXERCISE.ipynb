```python
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("spark").getOrCreate()
print("spark created successfully!!!")
```

```python
sc=spark.sparkContext
```

```python
logs = [
"Rahul login",
"Sneha login",
"Arjun login",
"Rahul purchase",
"Priya login",
"Sneha logout",
"Rahul login",
"Karan login",
"Arjun purchase",
"Sneha login",
"Rahul logout",
"Amit login",
"Priya purchase",
"Karan purchase",
"Sneha login",
"Rahul purchase",
"Meera login",
"Arjun login",
"Sneha purchase",
"Rahul login"
]
logs_rdd = sc.parallelize(logs)
logs_rdd.collect()
```

```python
sales = [
("Laptop",1,75000),
("Mouse",3,500),
("Keyboard",2,1500),
("Laptop",1,75000),
("Monitor",1,12000),
("Mouse",2,500),
("Keyboard",1,1500),
("Laptop",1,75000),
("Monitor",2,12000),
("Mouse",3,500),
("Laptop",1,75000),
("Keyboard",2,1500),
("Monitor",1,12000),
("Mouse",1,500),
("Laptop",1,75000)
]
sales_rdd = sc.parallelize(sales)
sales_rdd.collect()
```

```python
employees = [
("Rahul","IT",70000),
("Sneha","HR",60000),
("Arjun","IT",75000),
("Priya","Finance",80000),
("Karan","IT",50000),
("Amit","HR",58000),
("Meera","Finance",82000),
("Ravi","IT",72000),
("Neha","HR",61000),

("Vikram","Finance",90000)
]
emp_rdd = sc.parallelize(employees)
emp_rdd.collect()
```

Exercise 7 — All purchase actions

PART 2 — flatMap Practice

```python
sentences = sc.parallelize([
"big data spark",
"spark hadoop hive",
"python spark hadoop",
"spark streaming kafka"
])
```

Exercise 8 — Split sentences into word

```python
words = sentences.flatMap(lambda x: x.split())
words.collect()
```

Exercise 9 — Distinct technologies

```python
words.distinct().collect()
```

Exercise 10 — Word count

```python
words.map(lambda x: (x,1)) \
     .reduceByKey(lambda a,b: a+b) \
     .collect()
```

PART 3 — Sales Dataset
Exercise 11 — Revenue per row

```python
sales_rdd.map(lambda x: (x[0], x[1]*x[2])).collect()
```

Exercise 12 — Total revenue

```python
sales_rdd.map(lambda x: x[1]*x[2]).reduce(lambda a,b: a+b)
```

Exercise 13 — Distinct products

```python
sales_rdd.map(lambda x: x[0]).distinct().collect()
```

Exercise 14 — Quantity > 1

```python
sales_rdd.filter(lambda x: x[1] > 1).collect()
```

Exercise 15 — Total quantity per product

```python
sales_rdd.map(lambda x: (x[0], x[1])) \
         .reduceByKey(lambda a,b: a+b) \
         .collect()
```

Exercise 16 — Total revenue per product

```python
sales_rdd.map(lambda x: (x[0], x[1]*x[2])) \
         .reduceByKey(lambda a,b: a+b) \
         .collect()
```

Exercise 17 — Highest revenue product

```python
sales_rdd.map(lambda x: (x[0], x[1]*x[2])) \
         .reduceByKey(lambda a,b: a+b) \
         .sortBy(lambda x: x[1], ascending=False) \
         .take(1)
```

PART 4 — Employee Dataset
Exercise 18 — Employee names

```python
emp_rdd.map(lambda x: x[0]).collect()
```

Exercise 19 — Salary > 70000

```python
emp_rdd.filter(lambda x: x[2] > 70000).collect()
```

Exercise 20 — Distinct department

```python
emp_rdd.map(lambda x: x[1]).distinct().collect()
```

Exercise 21 — Total salary per department

```python
emp_rdd.map(lambda x: (x[1], x[2])) \
       .reduceByKey(lambda a,b: a+b) \
       .collect()
```

Exercise 22 — Average salary per department

```python
emp_rdd.map(lambda x: (x[1], (x[2],1))) \
       .reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1])) \
       .map(lambda x: (x[0], x[1][0]/x[1][1])) \
       .collect()
```

Exercise 23 — Highest paid employee

```python
emp_rdd.sortBy(lambda x: x[2], ascending=False).take(1)
```

PART 5 — Intermediate RDD
Exercise 24 — Employee count per department

```python
emp_rdd.map(lambda x: (x[1],1)) \
       .reduceByKey(lambda a,b: a+b) \
       .collect()
```

Exercise 25 — Sort by salary descending

```python
emp_rdd.sortBy(lambda x: x[2], ascending=False).collect()
```

Exercise 26 — Top 3 salaries

```python
emp_rdd.sortBy(lambda x: x[2], ascending=False).take(3)
```

Exercise 27 — (department, employee)

```python
emp_rdd.map(lambda x: (x[1], x[0])).collect()
```

Exercise 28 — Group employees by departmen

```python
emp_rdd.map(lambda x: (x[1], x[0])) \
       .groupByKey() \
       .mapValues(list) \
       .collect()
```

PART 6 — Advanced RDD
Exercise 29 — Department with highest total salary

```python
emp_rdd.map(lambda x: (x[1], x[2])) \
       .reduceByKey(lambda a,b: a+b) \
       .sortBy(lambda x: x[1], ascending=False) \
       .take(1)
```

Exercise 30 — Count + total salary per department

```python
emp_rdd.map(lambda x: (x[1], (1, x[2]))) \
       .reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1])) \
       .collect()
```

FINAL MINI PROJECT — Sales Analysis

1. Revenue per product

```python
revenue = sales_rdd.map(lambda x: (x[0], x[1]*x[2])) \
                   .reduceByKey(lambda a,b: a+b)
```

2. Top selling product (quantity)

```python
sales_rdd.map(lambda x: (x[0], x[1])) \
         .reduceByKey(lambda a,b: a+b) \
         .sortBy(lambda x: x[1], False) \
         .take(1)
```

3. Highest revenue product

```python
revenue.sortBy(lambda x: x[1], False).take(1)
```

4. Products sold more than 3 times

```python
sales_rdd.map(lambda x: (x[0], x[1])) \
         .reduceByKey(lambda a,b: a+b) \
         .filter(lambda x: x[1] > 3) \
         .collect()
```

5. Sort products by revenue

```python
revenue.sortBy(lambda x: x[1], False).collect()
```
