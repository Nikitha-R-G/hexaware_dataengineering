Part 2 — DataFrame Transformations

```python
from pyspark.sql.functions import *
```

 11. Convert order_date into date type

```python
orders_df = orders_df.withColumn(
    "order_date",
    to_date("order_date")
)
```

12. Add total_order_value column

```python
orders_enriched = orders_df.join(
    products_df.select("product_id","price"),
    "product_id"
)

orders_enriched = orders_enriched.withColumn(
    "total_order_value",
    col("price") * col("quantity")
)
```

13. Create stock_status column

```python
products_df = products_df.withColumn(
    "stock_status",
    when(col("stock_quantity") < 20, "Low Stock")
    .when(col("stock_quantity") < 50, "Medium")
    .otherwise("High")
)
```

 14. Create order_priority column

```python
orders_enriched = orders_enriched.withColumn(
    "order_priority",
    when(col("quantity") >= 50, "High")
    .when(col("quantity") >= 10, "Medium")
    .otherwise("Low")
)
```

15. Create expensive_product_flag

```python
products_df = products_df.withColumn(
    "expensive_product_flag",
    when(col("price") > 20000, "Yes").otherwise("No")
)
```

51. Rank products by revenue within category.

52. Rank suppliers by revenue within city

```python
%sql
SELECT
    supplier_name,
    supplier_city,
    SUM(bill_amount) AS revenue,
    RANK() OVER(
        PARTITION BY supplier_city
        ORDER BY SUM(bill_amount) DESC
    ) AS supplier_rank
FROM final_orders
GROUP BY supplier_name, supplier_city;
```

53. Top product per category using ROW_NUMBER

```python
%sql
SELECT *
FROM (
    SELECT
        product_name,
        category,
        SUM(bill_amount) AS revenue,
        ROW_NUMBER() OVER(
            PARTITION BY category
            ORDER BY SUM(bill_amount) DESC
        ) AS rn
    FROM final_orders
    GROUP BY product_name, category
) t
WHERE rn = 1;
```

74–76. MERGE Updates + Inserts

```python
%sql
MERGE INTO orders_target AS target
USING daily_updates AS source
ON target.order_id = source.order_id

WHEN MATCHED THEN
UPDATE SET
  target.product_id = source.product_id,
  target.supplier_id = source.supplier_id,
  target.order_date = source.order_date,
  target.quantity = source.quantity,
  target.order_status = source.order_status

WHEN NOT MATCHED THEN
INSERT (
order_id,
product_id,
supplier_id,
order_date,
quantity,
order_status
)
VALUES (
source.order_id,
source.product_id,
source.supplier_id,
source.order_date,
source.quantity,
source.order_status
);
```

80. Explain why MERGE is important.

MERGE is used when new data arrives continuously and we need to synchronize it with an existing table.

In real data pipelines, incoming data can contain:
New records that must be inserted
Existing records that must be updated
Corrected or changed information

Instead of running separate INSERT and UPDATE statements, MERGE performs both operations in a single atomic transaction.

Key Reasons for Using MERGE
1. Handles Incremental Data
2. Performs UPSERT Operation
3. Prevents Duplicate Records
4. ACID Transaction Support
5. Efficient for Large Data
6. Supports Real-Time Pipelines

Commonly used in CDC (Change Data Capture) and streaming pipelines.

Without MERGE:
Run UPDATE
Run INSERT
Manage logic manually

With MERGE:
One command handles update and insert safely.


87. Why Delta supports updates better —

Delta Lake supports updates efficiently because it provides ACID transactions, ensuring reliable and consistent data modifications. Unlike raw Parquet files, Delta maintains a transaction log that tracks all changes, enabling safe updates, deletes, and merges without rewriting entire datasets. It also supports concurrent operations, schema enforcement, and time travel, making data updates faster, safer, and easier to manage compared to Parquet.
