74. Merge Daily Updates into Target Table

```python
%sql
MERGE INTO visits_target AS target
USING daily_updates AS source
ON target.visit_id = source.visit_id

WHEN MATCHED THEN
UPDATE SET
    target.patient_id = source.patient_id,
    target.doctor_id = source.doctor_id,
    target.visit_date = source.visit_date,
    target.visit_status = source.visit_status,
    target.tests_count = source.tests_count

WHEN NOT MATCHED THEN
INSERT (
    visit_id,
    patient_id,
    doctor_id,
    visit_date,
    visit_status,
    tests_count
)
VALUES (
    source.visit_id,
    source.patient_id,
    source.doctor_id,
    source.visit_date,
    source.visit_status,
    source.tests_count
);
```

80. Why MERGE is Used for Incremental Loads ?
MERGE (UPSERT) is used because:

1.Updates existing records
2.Inserts new records automatically
3.Prevents duplicates
4. Maintains single source of truth
5. Supports CDC (Change Data Capture)
6. Efficient for daily batch ingestion
7. Delta Lake keeps full version history

Without MERGE you would need:

-> separate UPDATE
-> separate INSERT
-> duplicate checks
-> complex logic

MERGE performs all in one atomic ACID transaction.

If you update:
Read entire dataset
Modify rows
Rewrite full files
Replace old data manually
Slow + risky.

-> Delta Advantage
Delta Lake adds:
Transaction log (_delta_log)
Row-level operations
ACID transactions
Version control

So Spark only rewrites affected data files, not whole dataset.

Faster
Safer
Production-ready incremental load
