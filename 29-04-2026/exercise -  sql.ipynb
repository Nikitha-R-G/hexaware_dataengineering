```python
%sql
SELECT p.project_name,COUNT(e.emp_id) AS employee_count FROM projects p LEFT JOIN employees e ON p.department = e.department GROUP BY p.project_name;
```

Section 7 — Advanced Queries

36. Find top 3 highest paid employees in Engineering department.

```python
%sql
select emp_name,department,salary from employees where department="Engineering" order BY salary desc limit 3;
```

37. Find department with highest average salary.

```python
%sql
SELECT department,AVG(salary) AS avg_salary
FROM employees GROUP BY department ORDER BY avg_salary DESC
LIMIT 1;
```

38. Find city with highest number of employees.

```python
%sql
select city,count(*) as count_emp from employees group by city
order by count_emp desc limit(1);
```

39. Find employees whose salary is higher than department average.

```python
%sql
select emp_name,department,salary  from employees e where salary>(select avg(salary) from employees where department= e.department)
```

40. Find employees whose experience is highest in their department.

```python
%sql
select emp_name,department,experience  from employees e where experience=(select max(experience) from employees where department= e.department)
```

```python
%sql
CREATE OR REPLACE TABLE employee_salary_analysis AS
SELECT
    emp_name,
    city,
    department,
    salary,
    experience,

    CASE
        WHEN salary >= 95000 THEN 'High'
        WHEN salary >= 80000 THEN 'Medium'
        ELSE 'Low'
    END AS salary_category,

    AVG(salary) OVER (PARTITION BY department) AS department_avg_salary

FROM employees;
```
