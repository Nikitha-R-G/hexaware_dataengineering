Exercise 71 — Rank products by price

Exercise 73 — Row number by signup_date

```python
w = Window.orderBy("signup_date")
customers.withColumn("row_num", row_number().over(w)).show()
```

Exercise 74 — Rank customers by revenue

```python
rev = orders.join(customers,"customer_id") \
.join(order_items,"order_id") \
.join(products,"product_id") \
.withColumn("revenue", col("quantity")*col("price")) \
.groupBy("name").sum("revenue")

w = Window.orderBy(col("sum(revenue)").desc())

rev.withColumn("rank", rank().over(w)).show()
```

Exercise 75 — Dense rank

```python
rev.withColumn("dense_rank", dense_rank().over(w)).show()
```

Exercise 76 — Top customer per city

```python
w = Window.partitionBy("city").orderBy(col("sum(revenue)").desc())

orders.join(customers,"customer_id") \
.join(order_items,"order_id") \
.join(products,"product_id") \
.withColumn("revenue", col("quantity")*col("price")) \
.groupBy("city","name").sum("revenue") \
.withColumn("rank", rank().over(w)) \
.filter("rank=1").show()
```

Exercise 77 — Top product per category

```python
w = Window.partitionBy("category").orderBy(col("price").desc())

products.withColumn("rank", rank().over(w)) \
.filter("rank=1").show()
```

Exercise 78 — Running revenue by order date

```python
w = Window.orderBy("order_date")

orders.join(order_items,"order_id") \
.join(products,"product_id") \
.withColumn("revenue", col("quantity")*col("price")) \
.withColumn("running_total", sum("revenue").over(w)) \
.show()
```

Exercise 79 — Running total by produc

```python
w = Window.partitionBy("product_id").orderBy("order_id")

order_items.join(products,"product_id") \
.withColumn("revenue", col("quantity")*col("price")) \
.withColumn("running_total", sum("revenue").over(w)) \
.show()
```

Exercise 80 — Rank orders by payment

```python
w = Window.orderBy(col("amount").desc())

payments.withColumn("rank", rank().over(w)).show()
```

Final Capstone Mini Project

1. Join all tables

```python
final_df = customers.join(orders,"customer_id") \
    .join(order_items,"order_id") \
    .join(products,"product_id") \
    .join(payments,"order_id")
```

2. Calculate revenue

```python
from pyspark.sql.functions import sum

revenue_df = final_df.groupBy("customer_id") \
    .agg(sum("amount").alias("revenue"))
```

3. Find top customers

```python
revenue_df.orderBy("revenue", ascending=False).show()
```

4. Find top products

```python
final_df.groupBy("product_name") \
    .agg(sum("amount").alias("revenue")) \
    .orderBy("revenue", ascending=False) \
    .show()
```

5. Revenue per city

```python
final_df.groupBy("city") \
    .agg(sum("amount").alias("city_revenue")) \
    .show()
```

6. Rank customers by revenue

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank

windowSpec = Window.orderBy(final_df["amount"].desc())

ranked_df = final_df.withColumn(
    "rank",
    rank().over(windowSpec)
)
```
