```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("attendance.csv")
```

```python
# Convert datetime
df['clock_in'] = pd.to_datetime(df['clock_in'])
df['clock_out'] = pd.to_datetime(df['clock_out'])
```

```python
# Calculate work hours
df['work_hours'] = (
    df['clock_out'] - df['clock_in']
).dt.total_seconds() / 3600

```

```python
# Productivity score
df['productivity_score'] = (
    df['tasks_completed'] / df['work_hours']
)
```

```python
# Clean missing values
df = df.dropna()
```

```python
# Top performers
top = df.sort_values(
    by='productivity_score',
    ascending=False
).head(5)
```

```python
# Frequent absentees
absentees = df[df['work_hours'] < 4]

print("Top Performers")
print(top)

print("Frequent Absentees")
print(absentees)
```

```python
# Summary report
summary = df.groupby('employee_id')[
    ['work_hours','productivity_score']
].mean()

print(summary)
```
