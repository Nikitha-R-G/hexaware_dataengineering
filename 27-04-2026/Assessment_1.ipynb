Section 8 — When / Otherwise (71–75)
71. Patient age category

```python
patients = patients.withColumn(
    "age_category",
    when(col("age") < 30, "Young")
    .when(col("age") <= 45, "Adult")
    .otherwise("Senior")
).show()
```

72. Bill category

```python
bills = bills.withColumn(
    "bill_category",
    when(col("bill_amount") >= 1500, "High")
    .when(col("bill_amount") >= 1000, "Medium")
    .otherwise("Low")
).show()
```

73. Doctor fee category

```python
doctors = doctors.withColumn(
    "fee_category",
    when(col("consultation_fee") >= 1500, "Premium")
    .when(col("consultation_fee") >= 1000, "Standard")
    .otherwise("Basic")
).show()
```

Section 10 — Window Functions (86–95)
86. Rank patients by age within city

```python
from pyspark.sql.window import Window
w = Window.partitionBy("city").orderBy(col("age").desc())

patients.withColumn("rank", rank().over(w)).show()
```

87. Oldest patient per city

```python
patients.withColumn("rn", row_number().over(w))\
.filter(col("rn")==1).show()
```

88. Dense rank doctors by fee within specialization

```python
w2 = Window.partitionBy("specialization")\
           .orderBy(col("consultation_fee").desc())

doctors.withColumn("dense_rank",
                   dense_rank().over(w2)).show()
```

89. Highest fee doctor per specialization

```python
doctors.withColumn("rn",row_number().over(w2))\
.filter(col("rn")==1).show()
```

90. Top 2 oldest patients per city

```python
patients.withColumn("rn",row_number().over(w))\
.filter(col("rn")<=2).show()
```

91. Rank doctors by fee overall

```python
w3 = Window.orderBy(col("consultation_fee").desc())

doctors.withColumn("rank",rank().over(w3)).show()
```

92. Rank cities by patient count

```python
city_rank = patients.groupBy("city").count()

w4 = Window.orderBy(col("count").desc())

city_rank.withColumn("rank",rank().over(w4)).show()
```

93. Rank doctors by appointment count

```python
doc_count = appointments.groupBy("doctor_id").count()

doc_count.withColumn(
    "rank",
    rank().over(Window.orderBy(col("count").desc()))
).show()
```

94. Running total bill amount by payment mode

```python
w5 = Window.partitionBy("payment_mode")\
           .orderBy("bill_id")

bills.withColumn(
    "running_total",
    sum("bill_amount").over(w5)
).show()
```

95. Running appointment count by doctor

```python
w6 = Window.partitionBy("doctor_id")\
           .orderBy("appointment_date")

appointments.withColumn(
    "running_appt",
    count("*").over(w6)
).show()
```

10. Top billing patient

```python
spark.sql("""
SELECT p.name, SUM(b.bill_amount) total_bill
FROM patients p
JOIN appointments a ON p.patient_id=a.patient_id
JOIN bills b ON a.appointment_id=b.appointment_id
GROUP BY p.name
ORDER BY total_bill DESC
LIMIT 1
""").show()
```
