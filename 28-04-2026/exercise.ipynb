Section 7 — Data Cleaning
61. Remove duplicate rows

```python
df = df.dropDuplicates()
```

62. Check if there are null values

```python
from pyspark.sql.functions import col, sum

df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns]).show()
```

63. Replace null salary with 0

```python
df = df.fillna({"salary":0})
```

64. Drop rows where department is null

```python
df = df.dropna(subset=["department"])
```

65. Fill missing experience with average experience

```python
from pyspark.sql.functions import avg

avg_exp = df.select(avg("experience")).first()[0]
df = df.fillna({"experience":avg_exp})
```
